```python
#3
def top_k_words(text, k):
    lowtext=text.lower()
    a=',.:;?!@#$%^&*()'
    bez_a=""
    for i in lowtext:
        if i not in a:
            bez_a+=i
    words=bez_a.split()
    count_word={}
    for j in words:
        if j in count_word:
            count_word[j]+=1
        else:
            count_word[j]=1
    t=sorted(count_word.items(), key=lambda x:(-x[1],x[0]))
    r=[]
    for tt in t:
       r.append(tt)
    return r
text = "Кот, собака, кот! Мышь, собака, аист, собака."
print(top_k_words(text, 2))
```

```python
#9
def alternate_case_blocks(text, n):
    r=''
    for i in range(0,len(text), n):
        a=text[i:i+n]
        b=(i//n)+1
        if b%2==0:
            r+=a.upper()
        else:
            r+=a.lower()
    return r
message = "HelloWorldPythonCode"
result = alternate_case_blocks(message, 3)
print(result)
```

```python
#11
def common_unique_chars(s1, s2):
    r=""
    a=set()
    for i in s1:
        if i !=" ":
            if i in s2 and i not in a:
                r+=i
                a.add(i)
    return r
string1 = "a b1c d e2f a!c"
string2 = "x y c! a 2 f z"
result = common_unique_chars(string1, string2)

print(result)
```

```python
#16
def transform_list(nums):
    r=[]
    rr=[]
    for i in nums:
        if i>=0:
            r.append(i)
    for j in r:
        if j%2==0:
            rr.append(j**2)
        elif j%2!=0 and j>10:
            rr.append(sum(int(w) for w in str(j)))
        else:
            rr.append(j)
    return rr
numbers = [4, -5, 15, 2, 7, -10, 23]
print(transform_list(numbers))
```

```python
#22
def group_by_parity_and_sort(nums):
    n=[]
    nn=[]
    for i in nums:
        if i%2==0:
            nn.append(i)
        else:
            n.append(i)
    a=sorted(n)
    b=sorted(nn)
    return a,b
numbers = [15, 2, 8, 7, 1, 10, 4, 11]
print(f"Результат: {group_by_parity_and_sort(numbers)}")
```

```python
#2.3
def merge_dicts_sum(d1, d2):
    a1=d1.copy()
    for i,j in d2.items():
        if i in a1:
            a1[i]+=j
        else:
            a1[i]=j
    return a1
day_one = {'apple': 10, 'banana': 5, 'orange': 8}
day_two = {'banana': 12, 'kiwi': 4, 'apple': 3}
result = merge_dicts_sum(day_one, day_two)
print(result)
```

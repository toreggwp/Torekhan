```python
#16
def transform_list(nums):
    r=[]
    for i in nums:
        if i>=0:
            if i%2!=0 and i>=10:
                b=sum(int(j) for j in str(i))
                r.append(b)
            else:
                r.append(i**2)
```

```python
#18
def flatten_and_filter(lst):
    r=[]
    for i in lst:
        if i>=0:
            if i%4==0:
                if sum(int(j) for j in str(i))==1:
                    r.append(i)
    return (sorted(r))
```

```python
#22
def group_by_parity_and_sort(nums):
    n=[]
    nn=[]
    for i in nums:
        if i%2==0:
            n.append(i)
        else:
            nn.append(i)
    sn=sorted(n)
    snn=sorted(nn)
    return(sn,snn)
```

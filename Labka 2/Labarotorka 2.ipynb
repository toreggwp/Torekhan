```python
with open('orders.json','r',encoding='utf-8') as file:
    orders=json.load(file)
total_sum=0
total_pokupk=0
users_pokupk={}
user_order_counts={}
top_user=""
max_order_sum=0
for order in orders:
    total_sum+=order['total']

    user=order['user']
    if user in user_order_counts:
        user_order_counts[user]+=1
    else:
        user_order_counts[user]=1

    if order['total']>max_order_sum:
        max_order_sum=order['total']
        top_user=order['user']

    for item in order['items']:
        total_pokupk+=1

        if item in users_pokupk:
            users_pokupk[item]+=1
        else:
            users_pokupk[item]=1

popular_item=max(users_pokupk, key=users_pokupk.get)

tt=[
{"total_revenue":total_pokupk,

  "top_user": top_user,

  "most_popular_item":popular_item,

  "total_orders":len(users_pokupk), }
]

with open('summary.json','w',encoding='utf-8') as file:
    json.dump(tt,file,ensure_ascii=False,indent=2)
```

```python
import csv
import json

suspicious_tranzaksia=[]
suspicious_users=set()
users_counts={}
total_suspicious_tranzasia=0
userss=set()
with open('transactions.csv','r',encoding='utf-8') as file:
    a=csv.DictReader(file)
    for line in a:
        users=line['user_id']
        amount=int(line['amount'])
        userss.add(line['user_id'])
        if amount>500000:
            suspicious_tranzaksia.append(amount)
            total_suspicious_tranzasia+=amount

        if users in users_counts:
            users_counts[users]+=1
        else:
            users_counts[users]=1

for i,j in users_counts.items():
    if j>3:
        suspicious_users.add(i)

suspicious_users_list=list(suspicious_users)

with open('fraud_report.txt','w',encoding='utf-8') as txt_file:
    txt_file.write(f"Подозрительных транзакций:{len(suspicious_tranzaksia)}\n")
    txt_file.write(f"Подозрительных пользователей:{len(suspicious_users)}\n")
    txt_file.write(f"Список пользователей:{userss}\n")
    txt_file.write(f"Общая сумма подозрительных операций:{total_suspicious_tranzasia}\n")
with open('fraud_users.json','w',encoding='utf-8') as json_file:
    json.dump(suspicious_users_list, json_file, ensure_ascii=False, indent=4)
```

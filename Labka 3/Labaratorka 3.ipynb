```python
#2.1
def even_numbers(n):
    for i in range(2,n+1,2):
        if i==4:
            yield "кратно 4"
        elif i%2==0:
            yield i
for x in even_numbers(10):
    print(x)
```

```python
#2.2
words = ["кот", "машина", "арбуз", "дом"]
def filter_words(words):
    for i in words:
        if "а" in i:
            yield "c a"
        elif len(i)>3:
            yield i
for w in filter_words(words):
    print(w)
```

```python
#2.3
def infinite_numbers():
    i=1
    while True:
        if i%15==0:
            yield "FizzBuzz"
        elif i%3==0:
            yield "Fizz"
        elif i%5==0:
            yield "Buzz"
        else:
            yield i
        i+=1
a=infinite_numbers()
for _ in range(15):
    print(next(a))
```

```python
#2.4
def squares(n):
    for i in range(1,n+1):
        if i%2==0:
            yield "чётный квадрат"
        else:
            yield i**2
for x in squares(10):
    print(x)
```

```python
#4.1
def is_prime(x):
    if x<2:
        return False
    for i in range(2,int((x/2)+1)):
        if x%i==0:
            return False
    return True
def special_numbers(n):
    for j in range(1,n+1):
         if j%15==0:
             yield "FizzBuzz"
         elif j%3==0:
             yield "Fizz"
         elif j%5==0:
             yield "Buzz"
         elif is_prime(j):
             yield "простое"
         else:
             yield j
for x in special_numbers(15):
    print(x)
```

```python
#4.3
def process_numbers(numbers):
    a=list(filter(lambda x:x>0, numbers))
    b=((lambda y:y/2 if y%2==0 else y*3+1 )(j) for j in a)
    for t in b:
        yield t
numbers = [5, -2, 8, 0, -7, 3]
for x in process_numbers(numbers):
    print(x)
```
